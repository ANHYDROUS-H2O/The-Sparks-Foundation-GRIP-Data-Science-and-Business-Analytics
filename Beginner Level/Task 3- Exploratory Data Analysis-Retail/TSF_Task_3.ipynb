```python
df=df.drop(['Postal Code'],axis=1)
```

```python
plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category',data=df)
plt.xlabel('Sub-Category')
plt.xlabel('Category')
plt.show()
```

The State of California has the maximum number of purchases registered followed by New York

```python
plt.figure(figsize=(15,5))
sns.countplot(x=df['Sub-Category'])
plt.xticks(rotation=90)
plt.show()
```

Binders and Paper are bought more as they usually are in greater demand.

```python
plt.figure(figsize=(15,5))
sns.countplot(x=df['Discount'])
plt.xticks(rotation=90)
plt.show()
```

The discounts are majorly none or 20%.

```python
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
sns.heatmap(df.cov(),annot=True)
plt.show()
```

```python
fig,a=plt.subplots(nrows=2,ncols=2,figsize=(15,10));
sns.countplot(x=df['Segment'],ax=a[0][0])
sns.countplot(x=df['Region'],ax=a[0][1])
sns.countplot(x=df['Category'],ax=a[1][0])
sns.countplot(x=df['Ship Mode'],ax=a[1][1])
```

The maximum of the purchases are from west and east regions.

Consumers have the maximum number of sales purchases.

Office supplies are high in demand.

Shipping is majorly done from the Standard Class.

```python
plt.figure(figsize=(10,5))
sns.barplot(x=df['Sub-Category'],y=df['Profit'])
plt.xticks(rotation=90)
plt.show()
```

Although the sales for copiers are not much, they provide the maximum profit as copiers are usually expensive.

```python
plt.figure(figsize=(10,5))
sns.lineplot('Discount','Profit',data=df)
plt.show()
```

As seen above there is a massive drop in profit when the discount price is lowered to 50%. The optimum discount rate should be 0.1 or 10%

```python
catmode.plot(x='Category',y='Profit',kind='bar')
```

The Category Technology has the maximum profits.

```python
regmode.plot(x='Region',y='Profit',kind='bar')
```

The west regions have highest profit and the central regions have lowest profit.

# Top 5 Cities with Maximum Profit

```python
max_profit=pd.DataFrame(df.groupby(['City'])['Profit'].sum()).reset_index().sort_values('Profit',ascending=False)
max_profit.head(5)
```

New York City has the highest profit

# Top 5 cities with Minimum Profit

```python
min_profit=pd.DataFrame(df.groupby(['City'])['Profit'].sum()).reset_index().sort_values('Profit',ascending=True)
min_profit.head(5)
```

Philadelphia	has the lowest profit

# Segment wise Analysis

```python
df_segment=df.groupby(['Segment'])[['Sales','Discount','Profit']].mean()
df_segment
```

```python
df_segment.plot.pie(subplots=True,figsize=(20,15),autopct='%1.2f%%')
```

In terms of sales and profit, Home Office has slightly greater values.

The discounts of Consumer, Corporate and Home Office are approximately the same.
